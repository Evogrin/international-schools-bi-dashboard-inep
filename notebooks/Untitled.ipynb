```python
# 1. Selecionar só as colunas mais importantes
colunas_interesse = [
    'TP_DEPENDENCIA',    # tipo de escola (1 = federal, 2 = estadual, 3 = municipal, 4 = privada)
    'TP_LOCALIZACAO',    # 1 = urbana, 2 = rural
    'NO_UF',             # nome do estado
    'NO_MUNICIPIO',      # nome do município
    'QT_MAT_BAS',        # total de matrículas
    'QT_DOC_BAS',        # total de docentes
    'QT_TUR_BAS'         # total de turmas
]

# 2. Criar um DataFrame só com essas colunas (todas as escolas)
df_recorte = df[colunas_interesse]

# 3. Criar um DataFrame somente com escolas privadas
df_privadas = df_recorte[df_recorte['TP_DEPENDENCIA'] == 4]

# 4. Visualizar as 10 primeiras linhas de cada DataFrame
print("Todas as escolas:")
display(df_recorte.head(10))

print("Somente escolas privadas:")
display(df_privadas.head(10))



```

```python
caminho_base = r"C:\Users\Vitor\Desktop\Portifolio\international-school-data-viz-inep\data\microdados_censo_escolar_2023\dados"

# Salvar todas as escolas
df_recorte.to_csv(f"{caminho_base}\\todas_escolas.csv", index=False, sep=';', encoding='utf-8')
print("CSV de todas as escolas salvo com sucesso!")

# Salvar somente escolas privadas
df_privadas.to_csv(f"{caminho_base}\\escolas_privadas.csv", index=False, sep=';', encoding='utf-8')
print("CSV de escolas privadas salvo com sucesso!")
```
